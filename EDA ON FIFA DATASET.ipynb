# Data Visualization

```python
# Bar PLot
plt.figure(figsize=(8,6))
plt.bar(list(fifa['nationality'].value_counts()[0:10].keys()),
        list(fifa['nationality'].value_counts()[0:10]),
        color=["darkgreen","maroon","navy","fuchsia","goldenrod","purple","chocolate","olive","coral","crimson"])
plt.xticks(rotation=45, color='black')  # Rotate x-axis labels for better readability if needed
plt.show()        
```

```python
player_salary = fifa[['short_name','wage_eur']]
player_salary
```

```python
player_salary=player_salary.sort_values(by=['wage_eur'],ascending=False)
player_salary.head(10)
```

```python
# Bar Plot
plt.figure(figsize=(8,6))
plt.bar(list(player_salary['short_name'])[0:10],
        list(player_salary['wage_eur'])[0:10],
        color=["navy","limegreen","darkred","olive","crimson","brown","magenta","chocolate","gold","coral"])
plt.xticks(rotation=45, color='black')  # Rotate x-axis labels for better readability if needed
plt.show()
```
